```python
import numpy as np
import matplotlib.pyplot as plt

a = 0
k = 1/(2*np.pi)
t = np.linspace(0, 10*np.pi, 1000)
axe_x = np.arange(-4, 4)

r = a + k*t
x = r * np.cos(t)
y = r * np.sin(t)

plt.plot(x, y, 'r-')
plt.axis('equal')
plt.title('    ') 
plt.xlabel('x') 
plt.ylabel('y') 
plt.grid(True)
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='k')
plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = 0
k = 1/(2*np.pi)
t = np.linspace(0, 10*np.pi, 1000)
r = a+k*t

plt.subplot(111,projection='polar')
plt.title('    ')
plt.plot(t, r, color="r", linewidth=2)
plt.show()
```
